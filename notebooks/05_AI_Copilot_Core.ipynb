# AI Copilot for Panchayat Governance

## Phase 5 : AI Copilot Core

### Objective

In this notebook, we will build the core conversational engine of Sahayak AI.

By the end of this notebook, the AI will:

- Have a defined identity
- Follow a system prompt
- Maintain conversation history
- Generate contextual responses

```python
import google.generativeai as genai
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

# Configure Gemini
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# Load model
model = genai.GenerativeModel("gemini-2.5-flash")

print("✅ Gemini Model Loaded Successfully")
```

## Define the AI Identity

A good AI assistant should always know:

- Who it is
- What it does
- What it should never do

This is achieved using a system prompt.

```python
SYSTEM_PROMPT = """
You are Sahayak AI, an AI Copilot designed for Panchayat Governance in India.

Your primary responsibilities are:

1. Assist citizens with government schemes.
2. Help Panchayat officials draft documents.
3. Answer governance-related questions clearly.
4. Respond politely and professionally.
5. If you do not know an answer, say so honestly instead of inventing information.
6. Keep responses simple and easy to understand.
7. Support both Hindi and English where appropriate.
"""
```

```python
# Start a chat session
chat = model.start_chat(history=[])

print("✅ Chat Session Started")
```

```python
def ask_sahayak(user_query):
    """
    Send a message to Sahayak AI and return the response.
    """

    prompt = f"""
{SYSTEM_PROMPT}

User:
{user_query}

Assistant:
"""

    try:
        response = chat.send_message(prompt)
        return response.text

    except Exception as e:
        return f"Error: {e}"
```

```python
question = "What is PM Awas Yojana?"

answer = ask_sahayak(question)

print(answer)
```

```python
question = "Explain it in Hindi."

answer = ask_sahayak(question)

print(answer)
```

```python
while True:
    user_input = input("👤 You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("🤖 Sahayak AI: Thank you! Have a great day.")
        break

    response = ask_sahayak(user_input)
    print("\n🤖 Sahayak AI:")
    print(response)
    print("-" * 60)
```
