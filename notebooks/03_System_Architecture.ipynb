# AI Modules

### Chat Agent

Handles general conversation with users.

### Scheme Agent

Provides information about government welfare schemes.

### Eligibility Agent

Checks eligibility based on user inputs.

### Document Agent

Generates official letters, notices, certificates, and applications.

### Knowledge Agent

Searches the knowledge base using Retrieval-Augmented Generation (RAG).

### Voice Agent (Future)

Handles speech-to-text and text-to-speech functionality.

# Request Flow

1. User enters a query.
2. Frontend sends the request to FastAPI.
3. FastAPI authenticates the user.
4. AI Router identifies the query type.
5. Appropriate AI Agent processes the request.
6. If required, the agent retrieves information from the knowledge base.
7. The LLM generates a response.
8. The response is returned to the frontend.
